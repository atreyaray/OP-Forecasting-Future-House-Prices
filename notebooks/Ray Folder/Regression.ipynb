```python
import pandas as pd
import numpy as np
```

```python
# x
x = np.arange(0, 100, 0.1)
# y 
y = 3*x + 1
```

```python
from sklearn.model_selection import train_test_split
Xtr, Xtest, ytr, ytest = train_test_split(x,y, test_size = 0.3, random_state=3)
len(Xtest)
```

```python
noise = np.random.uniform(0, 1, size=700)
Xtr = Xtr + noise
Xtr
```

```python
from sklearn.linear_model import LinearRegression
Xtr = pd.DataFrame(Xtr)
ytr = pd.DataFrame(ytr)
Xtest = pd.DataFrame(Xtest)
lr = LinearRegression()
lr.fit(Xtr, ytr)
```

```python
ans = pd.DataFrame(np.array([30]))
y_pred = lr.predict(ans)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_pred, ytest))
```
