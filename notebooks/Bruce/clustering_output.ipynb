```python
import numpy as np
import pandas as pd
```

```python
building_type = 'two_room'
df = pd.read_csv(f"results/{building_type}/{building_type}_clusters.csv",index_col=[0])[["Postal code","Building type","label6"]]
df.head()
```

```python
df0 = pd.read_csv("cleaned.csv")

df0.head()
```

```python
joined = pd.merge(df, df0,  how='left', on = ['Postal code','Building type'])
joined.head()
```

```python
res_indices = joined.iloc[:,3:].columns
res_columns = range(0,6)
res_df = pd.DataFrame(index=res_indices,columns=res_columns)

def cluster_mean(df, cluster):
    return df[df['label6'] == cluster].mean()

for col in res_columns:
    res_df[col] = cluster_mean(joined, col)

res_df.head()
```

```python
### Regex to change quarters to datetime
import datetime as date

def quarter_to_datetime(row):
    dates = {"Q1": "1/1", "Q2":"1/4", "Q3":"1/7", "Q4":"1/10"}
    string = row.replace("*", "").split(" ")[0]
    year, quarter = string[:4], string[4:]
    datetime_str = dates[quarter] + "/" + year
    return date.datetime.strptime(datetime_str, '%d/%m/%Y')
    
    
res_df.index = res_df.index.to_series().apply(lambda row: quarter_to_datetime(row))
res_df.head()
```

```python
res_df.to_csv(f"results/{building_type}/{building_type}_cluster_means.csv",index=True)
df[["Postal code",'label6']].to_csv(f"results/{building_type}/{building_type}_cluster_dictionary.csv",index=False)
```
