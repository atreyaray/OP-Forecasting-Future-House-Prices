```python
import pandas as pd
```

```python
flats = pd.read_csv('../data/pxnet_flats.csv', encoding='unicode_escape')
terraced = pd.read_csv('../data/pxnet_terraced.csv', encoding='unicode_escape')
total = pd.read_csv('../data/pxnet_btype_total.csv', encoding='unicode_escape')
data = [flats, terraced, total]
```

```python
df = pd.concat(data, ignore_index=True)
df
```

```python
df_num = pd.DataFrame(df.iloc[:, 3:].apply(pd.to_numeric, errors="coerce"))
df_num[df.iloc[:, :3].columns] = df.iloc[:, :3]
cols = list(df.columns)[:3] + list(df.columns)[3:]
df_num = df_num[cols]
df_num
```

```python
missing_rows = 0
for i in range(492):
    if (df_num.iloc[i, :].isnull().any()):
        missing_rows += 1
missing_rows        
```
