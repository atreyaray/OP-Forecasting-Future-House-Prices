```python
import pandas as pd
```

```python
data_raw = pd.read_csv('../data/etuovi_raw.csv')
```

```python
data_raw[['Type','Raw_2']] = data_raw.Raw_Data.str.split("|",expand=True, n=1)
data_raw[["Raw_3", "Year"]] = data_raw.Raw_2.str.split("Vuosi", expand=True, n=1)
data_raw['Year_Built'] = data_raw['Year'].str[:4]
data_raw[["Raw_4", "Price_Iteration"]] = data_raw.Raw_3.str.split("Hinta", expand=True, n=1)
data_raw[["Price", "Size"]] = data_raw.Price_Iteration.str.split("Koko", expand=True, n=1)
data_raw["Size_m²"] = data_raw["Size"].str.replace(" m²","")
data_raw['Price_euro'] = data_raw['Price'].str.replace('€',"")

data_raw
```
