```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = None
ext = None
while not ext in ['csv','xlsx','json']:
    filename = input("Enter the name of the file to be read:\n")
    ext = filename.split('.')[-1]
    if not ext in ['csv','xlsx','json']:
        print('File extension \'{}\' is not supported. Please try again.'.format(ext))
        continue
    try:
        if ext == 'csv':
            df = pd.read_csv(filename)#, dtype="str", index_col=False)
        elif ext == 'xlsx':
            df = pd.read_excel(filename)#, dtype="str", index_col=False)
        elif ext == 'json':
            df = pd.read_json(filename)#, dtype="str", index_col=False)
    except:
        print("Error in reading the file \'{}\'. Please try again.".format(filename))
        ext = ''
```
